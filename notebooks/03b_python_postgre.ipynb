```python
import pandas as pd
from sqlalchemy import create_engine

# Read CSV
df = pd.read_csv(
    r"C:\Users\yoga\Desktop\Retail\data_raw\SampleSuperstore_clean.csv"
)

print("Rows before load:", len(df))

# PostgreSQL connection
engine = create_engine(
    "postgresql+psycopg2://postgres:Jamal1*@localhost:5432/retail_analytics_db"
)

# Write to PostgreSQL in chunks
df.to_sql(
    "stg_superstore",
    engine,
    schema="retail",
    if_exists="replace",
    index=False,
    chunksize=1000,
    method="multi"
)

print("SUCCESSFULLY LOADED TO POSTGRESQL")
```

```python
query = "SELECT COUNT(*) FROM retail.stg_superstore"

result = pd.read_sql(query, engine)

print(result)
```
